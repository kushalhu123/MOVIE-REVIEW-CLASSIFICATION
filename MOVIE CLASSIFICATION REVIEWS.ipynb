```python
#WE HAVE GOT A PRETTY DESCENT NUMBER OF STOP WORDS NOW LET US CLEAT THIS IN THE DOCUMENT
```

```python
def get_simple_tage(tag):
    if tag.startswith('J'):
        return wordnet.ADV
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
# NOW YOU HAVE TO CLEAN THE WORDS FOR EACH OF THE DOCUMENT
```

```python
lemma=WordNetLemmatizer()
```

```python
def clean_reviews(word):
    output_words=[]
    for each in word:
        if each not in stop:
            tagofthat=pos_tag([each])[0][1]
            getthetag=get_simple_tage(tagofthat)
            #HERE YOU HAVE DONE THE LEMMATIZATION
            clean_word=lemma.lemmatize(each,pos=getthetag)
            output_words.append(clean_word)
    return output_words          
```

```python
newdocuments=[(clean_reviews(doc),category) for doc,category in documents]
```

```python
# COLLECT ALL THE WORDS FROM ALL THE DOCUMENTS
training_data=newdocuments[0:1500]
testing_data=newdocuments[1500:]
```

```python
print(len(training_data))
print(len(testing_data))
```

```python
#GET ALL THE WORDS FROM THE TRAINING_DATA
alltwords=[]
for doc in training_data:
    alltwords+=doc[0]

    
```

```python
len(alltwords)
# SO YOU HAVE AROUNG 707025 OF CLEAN WORDS FROM ALL THE DOCUMENTS TRY TO GET MAXIMUM OF TOP 3000 WORDS FROM THIS
```

```python
# GET ALL THE WORDS 
```

```python
from nltk import FreqDist
import nltk
```

```python
freq=nltk.FreqDist(alltwords)
```

```python
features=[w[0] for w in freq.most_common(3000)]
```

```python
#NOW YOU GOT ALL THE FEATURES
```

```python
def get_the_feature_set(words):
    dic={}
    words=set(words)
    for w in features:
        dic[w]=w in words
    return dic
        
```

```python
ntraining_data=[(get_the_feature_set(doc),category) for doc,category in training_data]
```
